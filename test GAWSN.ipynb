# Framework cho bài toán


+ Bài toán kết hợp giữa GA + Prim kết hợp thêm chiến lược local search đặc biệt
    + Biểu diễn cá thể ( Chromosome Representation) :
        - Một Chromosome là 1 vector nhị phân có độ dài n+1 ( gồm số 0 và 1 ) trong đó n là số cụm thỏa mãn K - coverage giải quyết ở Phase 1 . Ý nghĩa của việc biểu diễn : Vị trí của số 0 thể hiện rằng cụm tương ứng với vị trí đó không được xét để kết nối , vị trí của số 1 thể hiện rằng cụm tương ứng với vị trí đó được xét để kết nối . Sau khi xác định vị trí các cụm được chọn để xét kết nối , ta thực hiện áp dụng thuật toán Prim xuất phát từ Base để tính số Sensor cần thiết
        - A chromosome is a vector of integers of size $n + 1 $, where $n$ denotes the number of coverage areas. We incremented the count by 1 to incorporate the Base station. Our research paper defines a chromosome as a significant binary sequence comprising 0s and 1s. Here, a 1 denotes the location of the associated coverage area for establishing the connecting line. Conversely, a 0 indicates the corresponding coverage area where the response is negated, thus disregarding any potential connection path.

    +  Genetic Operators :
        - Crossover ( lai ghép ) Đề xuất chiến lược lai ghép đặc biệt ( Heuristic crossover ) Chọn 2 cá thể cha mẹ (P1,P2) từ quần thể ban đầu pta sẽ tạo ra 1 cá thể con Childrent như sau . Nếu Childrent[i] = P1[i] (nếu P1[i]=P2[i] hoặc P1[i] khác P2[i] nhưng random.random()< fitness(P1) / ( fitness(P1)+ fitness(P2) )
        - Mutatation ( đột biến ) Chúng tôi đề xuất Heuristic Mutation . 1 cá thể thỏa mãn khi nó có đủ sensor cho việc kết nối . Đối với những cá thể này ta sẽ duyệt các vị trí có giá trị 0 và biến chúng thành 1 với xác suất (pro cho trước - theo thực nghiệm) . Mỗi lần có sự thay đổi , giảm (pro) 1 lượng 1/(2*n) để đảm bảo rằng không có quá nhiều sự biến đổi (vì nếu biến đổi quá nhiều dễ bị vi phạm ) . Tương tự với cá thể không thỏa mãn, cũng thực hiện chiến lược như trên nhưng thay vào biến đổi 0 thành 1, thì chúng ta biến đổi 1 thành 0
        - Local search : Với cá thể tốt nhất ( Best_indi - có giá trị fitness cao nhất ) . Chúng tôi sẽ cho duyệt qua tất cả các vị trí của Best_indi và thay lần lượt từng điểm có giá trị 0 về 1. Nếu tạo ra được 1 cá thể mới thỏa mãn thì đây sẽ là cá thể tốt nhất mới .
    + Evaluation :
        - Giá trị fitness của 1 cá thể được theo 1 chiến lược đặc biệt như sau :
        (The fitness value of an individual is determined according to a special strategy as follows:) 
            + Nếu cá thể ấy thỏa mãn bài toán (đủ số sensor) thì fitness của nó sẽ trả về số lượng target thỏa mãn K-coverage và K-connect
            + Đối với cá thể không thỏa mãn ( không đủ số sensor ) , giá trị fitness sẽ được tính theo 1 cách đặc biệt . Nó sẽ bằng 1 / số lượng sensor còn thiếu .  Để đảm bảo rằng, đối với các cá thể ko thỏa mãn , cá thể nào thiếu ít sensors hơn thì có fitness cao hơn
        - In our study, we retain unsatisfactory individuals within the population to preserve potentially beneficial genetic material for subsequent generations. Notably, individuals with fewer sensor deficiencies are assigned higher fitness values than those with greater deficiencies. To consistently meet the constraint, individuals of higher quality are assigned elevated fitness values.
            Cá thể X tốt hơn Y <=> fitness(X)>fitness(Y)
    +  Selection and Replacement :
        - Từ 1 quần thể Popu1 có số phần tử là N . Ta sẽ tạo ra 1 quần thể mới là new_popu1 với số lượng cá thể cũng là N bằng chiến lược heuristic đặc biệt :
        - Starting from a population denoted as 'Popu1' comprising N elements, we will generate a new population, labeled 'new_popu1,' also consisting of N individuals, employing a specialized heuristic strategy.
        
            Đầu tiên sắp xếp lại các cá thể trong Popu1 theo thứ tự giảm dần của hàm fitness
            (Initially, the individuals within 'Popu1' will be arranged in descending order based on their fitness function.)

            + Nếu random.random() < m_m thì :
                + Thực hiện chọn 1 cá thể bất kì trong tập Popu1 để đột biến nếu random.random() <m.m
                + Thực hiện chọn 2 cá thể bất kì trong tập Popu1 để lai ghép
            + Nếu không :
                + Thực hiện chọn 1 cá thể bất kì trong N/20 cá thể đầu tiên của Popu1 để đột biến nếu random.random() <m.m
                + Thực hiện chọn 2 cá thể bất kì trong tập N/10 cá thể đầu tiên của Popu1 để lai ghép
            Các cá thể mới được tạo ra và thêm vào new_popu1 cho đến khi nào |new_popu1| = N
        - Gộp new_popu1 và Popu1 lại được 1 tập mới là population gồm 2*N cá thể. Thực hiện chiến lược chọn lọc theo cách :
                + Chọn N/5 cá thể tốt nhất trong tập population
                + Chọn 4N/5 cá thể trong tập population bằng vòng quay roulete 
             Ta sẽ được 1 tập mới có N cá thể
             
             
             
             
Input :
    N : số lượng cá thể trong quần thể
    m_m : Hệ số đột biến
    n : số vùng coverage
    
Output : 
    res : The number of targets satisfy both k-coverage and

# Hàm ban đầu

```python
import math
import numpy as np
import networkx as nx
import random
import copy
def minConnect(sensors11, sensors22, r) :            # tính số lượng Sensors cần thiết để tạo đường kết nối cho 2 vùng sensor11 và sensor22
    sensors1 = copy.deepcopy(sensors11)
    sensors2 = copy.deepcopy(sensors22)
    numSensor = 0
    len_sensor = len(sensors1)
    dis = lambda s1, s2: math.sqrt( (s1[0]-s2[0])**2 + (s1[1]-s2[1])**2 )
    while(len_sensor > 0) :
        minDis = 1e9
        minIndex = 0
        for i in range(0, len(sensors2)) :
            if dis(sensors1[0], sensors2[i]) < minDis : 
                minDis = dis(sensors1[0], sensors2[i])
                minIndex = i
        sensors1.pop(0)
        sensors2.pop(minIndex)
        len_sensor -= 1
        numSensor += int((minDis-2*r)/(2*r)) + 1
    return numSensor


        
```

```python
def graphh(u,setpointt):            # tạo một đồ thị gồm các vùng được xét
    setpointx=[]
    for i in range(len(u)):
        if u[i]==1:
            setpointx.append(setpointt[i])
    x=[]
    for i in range(len(setpointx)-1):
        for j in range(i+1,len(setpointx)):
            x.append([i,j,minConnect(setpointx[i],setpointx[j],r)])
    return x
        
```

```python
def calculate_minimum_spanning_tree_length(edges):           # tính số sensors cần cho kết nối toàn vùng nhờ thuật toán Prim

    # Tạo đồ thị
    G = nx.Graph()

    # Thêm các cạnh vào đồ thị
    for edge in edges:
        G.add_edge(edge[0], edge[1], weight=edge[2])

    # Tìm cây khung nhỏ nhất sử dụng thuật toán Prim
    mst = nx.minimum_spanning_tree(G)
    # Tính tổng trọng số của các cạnh trong cây khung nhỏ nhất
    mst_length = sum(data['weight'] for _, _, data in mst.edges(data=True))

    # In trọng số và các cạnh tương ứng
    return mst_length

```

## Giải quyết Phase 1- đặt sensor thỏa mãn K - coverage

```python
OFFSET = 1.00000001
random.seed(2)

class Phase1:
    def __init__(self, WIDTH, HEIGHT, targets, R, nSENSOR, K, bX, bY):
        self.WIDTH = WIDTH
        self.HEIGHT = HEIGHT
        self.nTARGET = len(targets)
        self.R = R
        self.nSENSOR = nSENSOR
        self.K = K
        self.targets = targets
        self.sensors = []
        self.base = (bX, bY)
        self.coverScore = 0


    def PlaceSensors(self):
        intersectPoint = []
        intersecRegion = []
        for i in range(self.nTARGET):
            isTargetihaveNointersect = True
            for j in range(i+1, self.nTARGET):
                dis = math.sqrt(math.pow(self.targets[i][0]-self.targets[j][0], 2) + math.pow(self.targets[i][1]-self.targets[j][1], 2))
                if  dis <= self.R * 2 :
                    isTargetihaveNointersect = False
                    p1x = 0.5*(self.targets[i][0]+self.targets[j][0]) + 0.5*math.sqrt(2*(2*self.R*self.R/(dis*dis))-1)*(self.targets[j][1]-self.targets[i][1])
                    p1y = 0.5*(self.targets[i][1]+self.targets[j][1]) + 0.5*math.sqrt(2*(2*self.R*self.R/(dis*dis))-1)*(self.targets[i][0]-self.targets[j][0])

                    p2x = 0.5*(self.targets[i][0]+self.targets[j][0]) - 0.5*math.sqrt(2*(2*self.R*self.R/(dis*dis))-1)*(self.targets[j][1]-self.targets[i][1])
                    p2y = 0.5*(self.targets[i][1]+self.targets[j][1]) - 0.5*math.sqrt(2*(2*self.R*self.R/(dis*dis))-1)*(self.targets[i][0]-self.targets[j][0])

                    intersectPoint.append((p1x, p1y))
                    intersectPoint.append((p2x, p2y))
            if(isTargetihaveNointersect) and i < self.nTARGET : intersecRegion.append([i])

        for p in intersectPoint:
            regionP = []
            for i in range(self.nTARGET):
                dis = math.sqrt(math.pow(p[0]-self.targets[i][0], 2) + math.pow(p[1]-self.targets[i][1], 2))
                if dis <= self.R * OFFSET  : regionP.append(i)
            
            if regionP not in intersecRegion:
                intersecRegion.append(regionP)

        intersecRegion.sort(key=len, reverse=True)
        
        numSatisfyCover = 0
        sensorTargetNeed = [self.K] * self.nTARGET
        sensorRegion = intersecRegion[0].copy()
        sensorRegionNow = sensorRegion.copy()
        sensorPosInThisRegion = []
        sensorsPositionThatTheThingThisPhase1Want = []
        isCoverage = False
        sensorLeft = self.nSENSOR
        while (sensorLeft > 0) and not isCoverage:

            sensorPosRight = False
            sensorX = 0
            secsorY = 0
            centerX = 0
            centerY = 0
            for i in intersecRegion[0]:
                centerX += self.targets[i][0]
                centerY += self.targets[i][1]
            centerX /= len(intersecRegion[0])
            centerY /= len(intersecRegion[0])


            while not sensorPosRight:
                r = random.random() * self.R
                angle = random.random() * math.pi * 2
                sensorX = centerX + r * math.cos(angle) 
                sensorY = centerY + r * math.sin(angle) 
                sensorPosRight = True
                for i in intersecRegion[0]:
                    dis = math.sqrt((sensorX-self.targets[i][0])**2 + (sensorY-self.targets[i][1])**2)
                    if dis > self.R or sensorX < 0 or sensorX > self.WIDTH or sensorY < 0 or sensorY > self.HEIGHT:
                        sensorPosRight = False
                        break
            sensorPosInThisRegion.append((sensorX, sensorY))
            sensorLeft -= 1

            for i in intersecRegion[0].copy():
                sensorTargetNeed[i] -= 1
                if sensorTargetNeed[i] == 0:
                    for region in intersecRegion:
                        if i in region: region.remove(i)

            isCoverage = True
            for i in range(self.nTARGET):
                if sensorTargetNeed[i] != 0:
                    isCoverage = False
                    break
            intersecRegion.sort(key=len, reverse=True)
        
            sensorRegionNow = intersecRegion[0].copy()
            if sensorRegionNow != sensorRegion:
                sensorsPositionThatTheThingThisPhase1Want.append((sensorRegion.copy(), sensorPosInThisRegion.copy()))
                sensorRegion = sensorRegionNow.copy()
                sensorPosInThisRegion.clear()
        if len(sensorPosInThisRegion) > 0:
            sensorsPositionThatTheThingThisPhase1Want.append((sensorRegion, sensorPosInThisRegion))
        
        return sensorsPositionThatTheThingThisPhase1Want       
```

```python
def individual1(len_indi):                  # biểu diễn cá thể dạng nhị phân 
    new1=[]
    new1.append(1)
    indi_1 = np.random.random(len_indi)
    for i in range(1,len_indi):
        if indi_1[i]>=0.5:
            new1.append(1)
        else:
            new1.append(0)
    return new1
def population(po_size,len_indi):
    popu1=[]
    for _ in range(po_size):
        popu1.append(individual1(len_indi))
    return popu1
```

## Crossover , Mutate

```python
def crossover1(indd1,indd2):
    new=[]
    for i in range(len(indd1)):
        if indd1[i]==indd2[i]:
            new.append(indd1[i])
        else:
            ft1= list_indi[tuple(indd1)]
            ft2= list_indi[tuple(indd2)]
            ft12 = ft1+ft2
            if random.random() <  (ft1/(ft1+ft2)):
                new.append(indd1[i])
            else:
                new.append(indd2[i])
    return new
            
def mutate1(ind1,m_m):
    ind2 = copy.deepcopy(ind1)
    m_mm = copy.deepcopy(m_m)
    if check(ind2):
        for i in range(1,len(ind2)):
            if ind2[i]==0:
                if random.random()<m_mm:
                    m_mm -= 1/(2*len(ind1))
                    ind2[i]=1
    else:
        for i in range(1,len(ind2)):
            if ind2[i]==1:
                if random.random()<m_m:
                    ind2[i]=0
                    m_mm -=1/(2*len(ind1))
    return ind2
```

# Chọn điểm từ vòng quay Roulete

```python
def select_individuals(popu,num_selections,list_indi):
    selected_individuals = []
    total_fitness = sum(list_indi[tuple(popu[i])] for i in range(len(popu)))

        # Tính xác suất cho mỗi cá thể
    probabilities = [list_indi[tuple(popu[i])] / total_fitness for i in range(len(popu))]

        # Chọn một số ngẫu nhiên trong phạm vi từ 0 đến 1
    for _ in range(num_selections):

        random_number = np.random.random()

        # Chọn cá thể tương ứng
        cumulative_probability = 0
        for i, prob in enumerate(probabilities):
            cumulative_probability += prob
            if random_number <= cumulative_probability:
                selected_individuals.append(popu[i])
                break

    return selected_individuals


```
